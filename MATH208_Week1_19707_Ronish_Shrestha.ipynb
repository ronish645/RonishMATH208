```python
#Q.1
import random
import matplotlib.pyplot as plt

# Generating 500 random numbers
random_numbers = [random.uniform(-20, 20) for _ in range(500)]

# Calculate mean using formula (Total sum/no.)
mean = sum(random_numbers) / len(random_numbers)
print("Mean:", mean)

# Calculate median using formula
sorted_numbers = sorted(random_numbers)
n = len(sorted_numbers)
if n % 2 == 0:    #if n is even
    m1= (sorted_numbers[n//2 - 1])
    m2= (sorted_numbers[n//2])
    median=(m1+m2)/2
else:
    median = sorted_numbers[n//2]
print("Median:", median)

# Calculate standard deviation
variance = sum((x - mean) ** 2 for x in random_numbers) / len(random_numbers)
std_deviation = variance ** 0.5
print("Standard Deviation:", std_deviation)

# Plot histogram with custom color
plt.hist(random_numbers, bins=10, edgecolor='k', color='red')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Generate 500 random numbers with mean = 10 and standard deviation = 0.5
random_numbers = [random.gauss(10, 0.5) for _ in range(500)]

# Calculate mean
mean = sum(random_numbers) / len(random_numbers)
print("Mean:", mean)

# Calculate median
sorted_numbers = sorted(random_numbers)
n = len(sorted_numbers)
median = (sorted_numbers[n//2 - 1] + sorted_numbers[n//2]) / 2 if n % 2 == 0 else sorted_numbers[n//2]
print("Median:", median)

# Calculate standard deviation
variance = sum((x - mean) ** 2 for x in random_numbers) / len(random_numbers)
std_deviation = variance ** 0.5
print("Standard Deviation:", std_deviation)

# Plot histogram with red bars
plt.hist(random_numbers, bins=10, edgecolor='k', color='red')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#Q.3
import pandas as pd                        #for data manipulation
import matplotlib.pyplot as plt             #for creating plots.

# Data
data = {
    "Cause of Death": ["Heart Disease", "Malignant Neoplasms", "Stroke",
                       "Chronic Respiratory Disease", "Accidents",
                       "Diabetes", "Alzheimer’s", "Influenza/Pneumonia",
                       "Septicemia", "Nephritis/Nephrosis"],
    "Number of Deaths (x 10,000)": [63.2, 56.0, 13.7, 12.5, 12.2, 7.2, 7.2, 5.6, 3.4, 4.5]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Sort in descending order
df = df.sort_values(by="Number of Deaths (x 10,000)", ascending=False)

# Calculate cumulative percentage
df['Cumulative Percentage'] = df['Number of Deaths (x 10,000)'].cumsum() / df['Number of Deaths (x 10,000)'].sum() * 100

# Create horizontal Pareto chart
fig, ax = plt.subplots(figsize=(8, 6))

# Horizontal bar plot for Number of Deaths
ax.barh(df['Cause of Death'], df['Number of Deaths (x 10,000)'], color='b', alpha=0.7, label='Deaths')
ax.set_ylabel('Cause of Death')
ax.set_xlabel('Number of Deaths (x 10,000)')

# Create a second y-axis for Cumulative Percentage
ax2 = ax.twiny()
ax2.plot(df['Cumulative Percentage'], df['Cause of Death'], color='r', marker='o', label='Cumulative Percentage')
ax2.set_xlabel('Cumulative Percentage (%)')

plt.title('Horizontal Pareto Chart of Leading Causes of Death (2006)')

# Show the plot
plt.show()
```

```python
#Q.4
# Input Data
data = [11, 14, 15, 15, 16, 16, 17, 18, 19, 21, 25, 36,
        12, 14, 15, 15, 16, 16, 17, 18, 19, 21, 25, 39,
        13, 14, 15, 15, 16, 17, 17, 18, 20, 22, 26, 43,
        13, 14, 15, 15, 16, 17, 17, 18, 20, 22, 26, 46,
        13, 14, 15, 16, 16, 17, 17, 18, 20, 22, 27, 50,
        13, 14, 15, 16, 16, 17, 17, 19, 20, 23, 27, 54,
        13, 14, 15, 16, 16, 17, 18, 19, 20, 23, 29, 59,
        13, 15, 15, 16, 16, 17, 18, 19, 20, 23, 30, 67,
        14, 15, 15, 16, 16, 17, 18, 19, 21, 24, 31, 34]

# Sort the data
data_sorted = sorted(data)

# Calculate Median
n = len(data_sorted)
if n % 2 == 0:   #if n is even.
    median = (data_sorted[n//2 - 1] + data_sorted[n//2]) / 2
else:
    median = data_sorted[n//2]
print("Median:", median)

# Calculate Mode
frequency = {}
for item in data_sorted:
    frequency[item] = frequency.get(item, 0) + 1

max_freq = max(frequency.values())
mode = [k for k, v in frequency.items() if v == max_freq]
print("Mode:", mode)

# Calculate Q1 and Q3
q1_index = n // 4                #by formula
q3_index = 3 * (n // 4)          #by formula

Q1 = (data_sorted[q1_index - 1] + data_sorted[q1_index]) / 2
Q3 = (data_sorted[q3_index - 1] + data_sorted[q3_index]) / 2
print("Q1 (First Quartile):", Q1)
print("Q3 (Third Quartile):", Q3)

# Calculate P10 and P95.  By formula
p10_index = n // 10
p95_index = 95 * (n // 100)

P10 = data_sorted[p10_index - 1]
P95 = data_sorted[p95_index - 1]
print("P10 (Percentile 10):", P10)
print("P95 (Percentile 95):", P95)
```
